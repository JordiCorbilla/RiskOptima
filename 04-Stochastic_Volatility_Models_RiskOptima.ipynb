# Stochastic Volatility Models

## Workflow overview

This notebook visualizes stochastic volatility model behaviour. The objective is to show how simulated paths can represent changing volatility regimes, jumps, and non-constant risk through time. These models are useful context for option pricing and stress testing because real markets rarely follow constant-volatility assumptions.


```python
import numpy as np
import matplotlib.pyplot as plt
from riskoptima import RiskOptima
print(RiskOptima.VERSION)
```

```python
def plot_simulations():
        """
        Runs simulations for all three models and plots their paths.
        """
        S_HW, sigma_HW = RiskOptima.simulate_hull_white()
        S_Heston, sigma_Heston = RiskOptima.simulate_heston()
        F_SABR, sigma_SABR = RiskOptima.simulate_sabr()

        plt.figure(figsize=(12, 6))
        plt.plot(S_HW, label="Hull-White Asset Price", linestyle='dashed')
        plt.plot(S_Heston, label="Heston Asset Price", linestyle='dotted')
        plt.plot(F_SABR, label="SABR Forward Price", linestyle='solid')
        plt.legend()
        plt.title("Simulated Price Paths for Different Stochastic Volatility Models")
        plt.xlabel("Time Steps")
        plt.ylabel("Price")
        plt.show()

        plt.figure(figsize=(12, 6))
        plt.plot(sigma_HW, label="Hull-White Volatility", linestyle='dashed')
        plt.plot(sigma_Heston, label="Heston Volatility", linestyle='dotted')
        plt.plot(sigma_SABR, label="SABR Volatility", linestyle='solid')
        plt.legend()
        plt.title("Simulated Volatility Paths for Different Stochastic Volatility Models")
        plt.xlabel("Time Steps")
        plt.ylabel("Volatility")
        plt.show()
```

```python
plot_simulations()
```
